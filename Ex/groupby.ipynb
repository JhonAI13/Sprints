```python
import pandas as pd
import numpy as np

data = {
    'cliente': ['Ana', 'Bruno', 'Carlos', 'Ana', 'Carlos', 'Bruno', 'Ana', 'Bruno', 'Carlos', 'Ana'],
    'categoria': ['A', 'B', 'A', 'B', 'B', 'A', 'A', 'B', 'A', 'B'],
    'valor_compra': [200, 150, 300, 100, 250, 400, 500, 350, 600, 50],
    'data_compra': pd.date_range(start='2024-01-01', periods=10, freq='D')
}

df = pd.DataFrame(data)
df
```

Agrupe o total gasto por cada cliente.

Calcule quantas compras cada cliente realizou.

Agrupe o total gasto por categoria de produto.

Calcule o valor m√©dio das compras para cada cliente.

Identifique a compra mais cara e a compra mais barata de cada cliente.

Determine o primeiro e o √∫ltimo dia de compra de cada cliente.

Crie uma nova coluna no DataFrame que indique o total de compras realizadas por cada cliente.

```python
# Agrupe o total gasto por cada cliente.
df.groupby('cliente')['valor_compra'].sum().to_frame()
```

```python
# Calcule quantas compras cada cliente realizou.
df.groupby('cliente')['cliente'].value_counts()
df['cliente'].value_counts()
```

```python
# Agrupe o total gasto por categoria de produto.
df.groupby('categoria')['valor_compra'].sum()
```

```python
# Calcule o valor m√©dio das compras para cada cliente.
df.groupby('cliente')['valor_compra'].mean()
```

```python
# Identifique a compra mais cara e a compra mais barata de cada cliente.
df.groupby('cliente')['valor_compra'].agg(valor_max='max', valor_min='min')

```

```python
# Determine o primeiro e o √∫ltimo dia de compra de cada cliente.
df.groupby('cliente')['data_compra'].agg(data_max='max', data_min='min')
```

```python
# Crie uma nova coluna no DataFrame que indique o total de compras realizadas por cada cliente.
agrupado = df.groupby('cliente')['valor_compra'].sum()
pd.merge(df, agrupado, on='cliente')
```
